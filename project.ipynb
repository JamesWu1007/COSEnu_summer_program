```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

#  vz 
vz = np.linspace(0, 1, 8)

#  g_a  Gaussian 
def g_a(v, v0=1.0, sigma=0.6):
    exponant = (v - v0) ** 2 / (2.0 * sigma ** 2)
    N = sigma * np.sqrt(np.pi / 2.0) * (erf((1.0 + v0) / sigma / np.sqrt(2.0)) + erf((1.0 - v0) / sigma / np.sqrt(2.0)))
    return np.exp(-exponant) / N

#  g_a C
#def g_a(vz, C=2):
#    return np.full_like(vz, C)

#  g_b  Gaussian 
def g_b(v, v0=1.0, sigma=0.5):
    exponant = (v - v0) ** 2 / (2.0 * sigma ** 2)
    N = sigma * np.sqrt(np.pi / 2.0) * (erf((1.0 + v0) / sigma / np.sqrt(2.0)) + erf((1.0 - v0) / sigma / np.sqrt(2.0)))
    return np.exp(-exponant) / N

# 
t_transition = 0.0  # 
transition_width = 200.0  # 
initial_bee = 0.5  # t=0  alpha 
final_bee = 1.2  #  alpha 
bee_diff = final_bee - initial_bee

#  alpha 
def alpha(t):
    sigmoid = 1.0 / (1.0 + np.exp(-(t - t_transition) / transition_width))
    return initial_bee + bee_diff * sigmoid

#  G(vz) 
time_points = np.linspace(0, 1300, 11)  # 

plt.figure(figsize=(14, 6))

#  G(vz) 
plt.subplot(1, 2, 1)
for t in time_points:
    alpha_t = alpha(t)
    G = g_a(vz) - alpha_t * g_b(vz)
    plt.plot(vz, G, label=f'$t = {t}, \\alpha(t) = {alpha_t:.2f}$')

plt.axhline(0, color='gray', linestyle='--')
plt.title('$G(v_z) = g_a(v_z) - \\alpha(t) \\times g_b(v_z)$')
plt.xlabel('$v_z$')
plt.ylabel('$G(v_z)$')
plt.legend()
plt.grid(True)

#  alpha 
time_points_alpha = np.linspace(0, 1300, 500)  # 
alpha_values = alpha(time_points_alpha)

#  alpha 
plt.subplot(1, 2, 2)
plt.plot(time_points_alpha, alpha_values, label='$\\alpha(t)$', color='blue')

plt.title('$\\alpha(t)$ vs Time')
plt.xlabel('Time')
plt.ylabel('$\\alpha(t)$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
